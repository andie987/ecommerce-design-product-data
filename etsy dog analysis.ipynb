````python
from scholarly import scholarly

def search_psych_literature(keywords, max_results=5):
    """
    Searches Google Scholar for the specified keywords and extracts relevant information.

    Args:
        keywords (list): List of keywords to search for.
        max_results (int): Maximum number of articles to fetch per keyword.

    Returns:
        list: A list of dictionaries containing article details.
    """
    results = []
    for keyword in keywords:
        print(f"Searching for: {keyword}")
        search_query = scholarly.search_pubs(query=keyword)  # Adjusted for new API usage
        for i, article in enumerate(search_query):
            if i >= max_results:
                break
            try:
                results.append({
                    "Title": article['bib']['title'],
                    "Authors": article['bib'].get("author", "N/A"),
                    "Publication": article['bib'].get("journal", "N/A"),
                    "Year": article['bib'].get("pub_year", "N/A"),
                    "Abstract": article['bib'].get("abstract", "N/A"),
                    "Link": article['pub_url'] if "pub_url" in article else "N/A"
                })
            except Exception as e:
                print(f"Error processing article: {e}")
    return results

# Define keywords for search
keywords = [
    "decision-making overload effect psychology",
    "anchoring pricing psychology",
    "consumer emotional triggers buying behavior",
    "choice architecture and consumer decisions"
]

# Fetch articles
articles = search_psych_literature(keywords, max_results=5)

# Display results
for article in articles:
    print(article)
````

````python
from collections import Counter
import re

# Function to extract keywords from text
def extract_keywords(text):
    if isinstance(text, str):
        # Remove special characters and split by whitespace
        words = re.findall(r'\b\w+\b', text.lower())
        # Filter out common stopwords (can be expanded)
        stopwords = set(['the', 'and', 'of', 'in', 'for', 'on', 'with', 'to', 'a', 'is', 'this', 'that'])
        keywords = [word for word in words if word not in stopwords and len(word) > 2]
        return keywords
    return []

# Extract keywords from title and description
market_data['keywords'] = market_data['title'].apply(extract_keywords) + market_data['description'].apply(extract_keywords)

# Flatten keyword list and calculate frequencies (supply metric)
all_keywords = [keyword for sublist in market_data['keywords'] for keyword in sublist]
keyword_supply = Counter(all_keywords)

# Group by keyword and calculate average popularity score (demand metric)
keyword_demand = market_data.explode('keywords').groupby('keywords')['popularity_score'].mean()

# Combine supply and demand metrics into a DataFrame
market_analysis = pd.DataFrame({
    'Supply': pd.Series(keyword_supply),
    'Demand': keyword_demand
}).dropna()

# Add supply-demand ratio column
market_analysis['Demand_Supply_Ratio'] = market_analysis['Demand'] / market_analysis['Supply']

# Sort by demand-supply ratio for analysis
market_analysis_sorted = market_analysis.sort_values(by='Demand_Supply_Ratio', ascending=False)

pd.set_option('display.max_rows', None) 
pd.set_option('display.max_columns', None)  
````

````python
import numpy as np
import pandas as pd
file_path = 'Etsy Product Scraper.csv'
data = pd.read_csv(file_path)

# Fill missing values in 'Number_of_reviews' with 0 for simplicity
data['Number_of_reviews'] = data['Number_of_reviews'].fillna(0)

# Convert the 'Price' column to numeric, forcing errors to NaN
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

# Fill NaN or invalid prices with a default value (e.g., 1 to avoid zero or negative values)
data['Adjusted_Price'] = data['Price'].fillna(1)

# Define new parameters for the updated formula
base_factor = 100  # Baseline constant to simulate base traffic
low_price_factor = 1.5
mid_price_factor = 1.0
high_price_factor = 0.5

# Define price range segmentation
def price_factor(price):
    if price < 5:
        return low_price_factor
    elif 5 <= price <= 15:
        return mid_price_factor
    else:
        return high_price_factor

# Apply the price factor function
data['Price_Factor'] = data['Adjusted_Price'].apply(price_factor)

# Calculate new estimated sales
data['Estimated_Sales'] = (
    (np.log(data['Number_of_reviews'] + 1) + base_factor) *  # Non-linear review weight with base factor
    (data['Rating'] ** 2) *  # Exponential impact of rating
    (data['Price_Factor'] / data['Adjusted_Price'])  # Adjusted by price and its factor
)
````

````python
# Calculate the popularity score
data['Popularity_Score'] = (
    np.log(data['Estimated_Sales'] + 1) *
    data['Rating'] *
    np.sqrt(data['Number_of_reviews'] + 1) /
    data['Price']
)

````

````python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Ensure the sentiment analyzer is available
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

# Step 1: Preprocess and Analyze Sentiment of Product Names
# Apply sentiment analysis to each product name
data['Sentiment_Scores'] = data['Name'].apply(lambda name: sia.polarity_scores(name))

# Extract individual sentiment components (positive, neutral, negative, and compound scores)
data['Positive_Sentiment'] = data['Sentiment_Scores'].apply(lambda x: x['pos'])
data['Neutral_Sentiment'] = data['Sentiment_Scores'].apply(lambda x: x['neu'])
data['Negative_Sentiment'] = data['Sentiment_Scores'].apply(lambda x: x['neg'])
data['Compound_Sentiment'] = data['Sentiment_Scores'].apply(lambda x: x['compound'])

# Step 2: Aggregate Sentiment Scores by Keywords
# For each keyword, calculate average sentiment scores
keyword_sentiment = pd.DataFrame()
for keyword in vectorizer.get_feature_names_out():
    keyword_sentiment.loc[keyword, 'Average_Positive'] = data[data['Name'].str.contains(keyword, case=False, na=False)]['Positive_Sentiment'].mean()
    keyword_sentiment.loc[keyword, 'Average_Neutral'] = data[data['Name'].str.contains(keyword, case=False, na=False)]['Neutral_Sentiment'].mean()
    keyword_sentiment.loc[keyword, 'Average_Negative'] = data[data['Name'].str.contains(keyword, case=False, na=False)]['Negative_Sentiment'].mean()
    keyword_sentiment.loc[keyword, 'Average_Compound'] = data[data['Name'].str.contains(keyword, case=False, na=False)]['Compound_Sentiment'].mean()

# Sort by Average_Compound Sentiment
keyword_sentiment = keyword_sentiment.sort_values(by='Average_Compound', ascending=False)

# Display the sentiment scores for each keyword
display(keyword_sentiment)

````

````python
import pandas as pd
import networkx as nx
from sklearn.feature_extraction.text import CountVectorizer
from scipy.spatial.distance import cosine
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'Etsy Product Scraper.csv'  # Update with your file path if needed
data = pd.read_csv(file_path)

# Preprocessing: Extract keywords and compute similarity
def extract_keywords(names):
    vectorizer = CountVectorizer(stop_words='english')
    keyword_matrix = vectorizer.fit_transform(names)
    keywords = vectorizer.get_feature_names_out()
    return keyword_matrix, keywords

# Create graph based on keyword similarity
def create_optimized_graph(data, keyword_matrix, threshold=0.2, sample_size=200):
    G = nx.Graph()
    sampled_data = data.sample(min(sample_size, len(data)), random_state=42)
    sampled_matrix = keyword_matrix[sampled_data.index]
    sellers = sampled_data['Seller'].tolist()

    # Add nodes (sellers)
    for seller in sellers:
        G.add_node(seller)

    # Compute similarity for a sample of products
    for i in range(sampled_matrix.shape[0]):
        for j in range(i + 1, sampled_matrix.shape[0]):
            vec_i = sampled_matrix[i].toarray().flatten()
            vec_j = sampled_matrix[j].toarray().flatten()
            similarity = 1 - cosine(vec_i, vec_j)
            if similarity >= threshold:
                seller_i = sampled_data.iloc[i]['Seller']
                seller_j = sampled_data.iloc[j]['Seller']
                if seller_i != seller_j:
                    G.add_edge(seller_i, seller_j, weight=similarity)
    return G

# Extract keywords and create the graph
names = data['Name'].fillna('').tolist()
keyword_matrix, keywords = extract_keywords(names)
optimized_graph = create_optimized_graph(data, keyword_matrix, threshold=0.2, sample_size=200)

# Plot the optimized graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(optimized_graph, seed=42)
nx.draw(
    optimized_graph,
    pos,
    with_labels=True,
    node_size=700,
    font_size=10,
    alpha=0.8,
    edge_color='gray'
)
plt.title("Optimized Competitive Seller Relationship Graph")
plt.show()
````

````python
#The visualization reveals that:

#With Keywords: The majority of products have low popularity scores, but the group also includes higher-scoring products, creating a longer tail.
#Without Keywords: This group has very few products, and their scores are concentrated within a narrower range.
#Observations:
#The skewed distribution and imbalance make direct statistical tests less reliable.
#The presence of outliers in both groups may also contribute to instability.
#Next Steps:
#Use bootstrapping to estimate the mean popularity score for both groups with confidence intervals.
#Provide a clearer measure of whether using keywords meaningfully improves popularity.

import numpy as np

# Bootstrapping function to calculate confidence intervals
def bootstrap_mean(data, num_samples=1000, ci=95):
    boot_means = [np.mean(np.random.choice(data, size=len(data), replace=True)) for _ in range(num_samples)]
    lower_bound = np.percentile(boot_means, (100 - ci) / 2)
    upper_bound = np.percentile(boot_means, 100 - (100 - ci) / 2)
    return np.mean(boot_means), lower_bound, upper_bound

# Perform bootstrapping for both groups
mean_with_keywords, lower_with, upper_with = bootstrap_mean(group_with_keywords.dropna())
mean_without_keywords, lower_without, upper_without = bootstrap_mean(group_without_keywords_cleaned)

# Display results
(mean_with_keywords, lower_with, upper_with), (mean_without_keywords, lower_without, upper_without)
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Define simulation parameters
num_days = 365 * 2  # Simulate two years of data
start_date = '2023-01-01'
date_range = pd.date_range(start=start_date, periods=num_days, freq='D')

# Define regions and their characteristics
regions = ['Northeast', 'South', 'Midwest', 'West']
region_characteristics = {
    'Northeast': {'base_inventory': 400, 'price_sensitivity': 1.1, 'marketing_impact': 1.2},
    'South': {'base_inventory': 500, 'price_sensitivity': 0.9, 'marketing_impact': 1.1},
    'Midwest': {'base_inventory': 300, 'price_sensitivity': 1.0, 'marketing_impact': 0.9},
    'West': {'base_inventory': 450, 'price_sensitivity': 1.2, 'marketing_impact': 1.3}
}

# Define products and their characteristics
products = ['DogSticker_A', 'DogSticker_B', 'DogSticker_C']
product_characteristics = {
    'DogSticker_A': {'popularity': 1.0, 'base_price': 10.0},
    'DogSticker_B': {'popularity': 0.9, 'base_price': 12.0},
    'DogSticker_C': {'popularity': 1.2, 'base_price': 15.0}
}

# Define retailers and their characteristics
retailers = ['Retailer_X', 'Retailer_Y', 'Retailer_Z']
retailer_characteristics = {
    'Retailer_X': {'market_share': 0.5, 'price_adjustment': -0.1},
    'Retailer_Y': {'market_share': 0.3, 'price_adjustment': -0.15},
    'Retailer_Z': {'market_share': 0.2, 'price_adjustment': -0.05}
}

# Retailer-specific supply chain disruptions
def retailer_supply_disruptions(day, retailer):
    """Simulate retailer-specific supply chain disruptions."""
    disruption_days = np.random.choice(range(num_days), size=int(num_days * 0.05), replace=False)
    if day in disruption_days:
        return 0.5  # 50% reduction in inventory
    return 1.0

# Adjust market share based on price wars
def adjust_market_shares(base_market_share, price_difference):
    """Adjust market shares dynamically based on price competition."""
    sensitivity = 0.2  # Sensitivity to price changes
    return base_market_share * (1 + sensitivity * price_difference)

# Simulate marketing synergies
def marketing_synergies(day, region, retailer):
    """Model combined effects of multiple campaigns across regions and retailers."""
    campaign_days = np.random.choice(range(num_days), size=int(num_days * 0.1), replace=False)
    synergy_effect = 1.0
    if day in campaign_days:
        synergy_effect *= region_characteristics[region]['marketing_impact'] * 1.1
    return synergy_effect

# Initialize an empty list to store all data
all_data = []

# Generate data for each region, product, and retailer
for region in regions:
    for product in products:
        for retailer in retailers:
            # Extract characteristics
            base_inventory = region_characteristics[region]['base_inventory']
            price_sensitivity = region_characteristics[region]['price_sensitivity']
            marketing_impact = region_characteristics[region]['marketing_impact']
            popularity = product_characteristics[product]['popularity']
            base_price = product_characteristics[product]['base_price']
            price_adjustment = retailer_characteristics[retailer]['price_adjustment']
            base_market_share = retailer_characteristics[retailer]['market_share']
            
            # Adjusted price and market share
            adjusted_price = base_price + price_adjustment * base_price
            price_difference = (base_price - adjusted_price) / base_price
            adjusted_market_share = adjust_market_shares(base_market_share, price_difference)
            
            # Base sales trend
            trend = 200 / (1 + np.exp(-0.01 * (np.arange(num_days) - 365)))  # Non-linear growth
            regional_seasonal_pattern = (
                50 * np.sin(2 * np.pi * np.arange(num_days) / 365)
            )
            weekly_pattern = 15 * np.sin(2 * np.pi * (date_range.dayofweek / 7)) + 20  # Weekly seasonality
            
            # Random daily noise
            daily_random_noise = np.random.normal(0, 20, num_days)
            
            # Marketing synergy effects
            marketing_effect = np.array([
                marketing_synergies(day, region, retailer) for day in range(num_days)
            ])
            
            # Base sales
            base_sales = trend + regional_seasonal_pattern + weekly_pattern + daily_random_noise
            adjusted_sales = base_sales * popularity * marketing_effect * adjusted_market_share
            adjusted_sales = np.maximum(adjusted_sales, 0)  # Ensure non-negative sales
            
            # Apply supply chain disruptions and inventory constraints
            inventory_levels = []
            sales_history = []
            for day in range(num_days):
                inventory_factor = retailer_supply_disruptions(day, retailer)
                inventory = base_inventory * inventory_factor
                sales = min(adjusted_sales[day], inventory)  # Constrain by inventory
                inventory_levels.append(inventory)
                sales_history.append(sales)
            
            # Create a DataFrame for the region-product-retailer combination
            region_product_data = pd.DataFrame({
                'Date': date_range,
                'Region': region,
                'Product': product,
                'Retailer': retailer,
                'Sales': sales_history,
                'Inventory': inventory_levels,
                'Price': adjusted_price,
                'MarketShare': adjusted_market_share,
                'MarketingEffect': marketing_effect
            })
            all_data.append(region_product_data)

# Combine all data into a single DataFrame
final_data = pd.concat(all_data, ignore_index=True)

# Save to CSV
final_data.to_csv('simulated_sales_with_price_wars_and_synergies.csv', index=False)

# Plot sales for a specific product and retailer in a region
for retailer in retailers:
    product_data = final_data[
        (final_data['Region'] == 'Northeast') & 
        (final_data['Product'] == 'DogSticker_A') & 
        (final_data['Retailer'] == retailer)
    ]
    plt.plot(product_data['Date'], product_data['Sales'], label=retailer)

plt.title('DogSticker_A Sales in Northeast by Retailer (Price Wars and Synergies)')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Display a sample of the data
final_data.head()
````

````python
import pandas as pd

# Load the uploaded CSV file to explore its contents
file_path = 'Etsy Product Scraper.csv'
etsy_data = pd.read_csv(file_path)


````

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import networkx as nx

# Extract product names for keyword-based similarity
product_names = etsy_data['Name'].astype(str)

# Use TF-IDF to vectorize product names
vectorizer = TfidfVectorizer(stop_words='english', max_features=500)
tfidf_matrix = vectorizer.fit_transform(product_names)

# Compute cosine similarity between products based on their names
similarity_matrix = cosine_similarity(tfidf_matrix)

# Threshold for creating edges (e.g., similarity > 0.3)
threshold = 0.3

# Create a graph where nodes represent products and edges represent similarity
graph = nx.Graph()

# Add nodes with attributes
for idx, row in etsy_data.iterrows():
    graph.add_node(idx, seller=row['Seller'], rating=row['Rating'], 
                   price=row['Price'], popularity=row['Popularity_Score'])

# Add edges based on the similarity threshold
for i in range(len(similarity_matrix)):
    for j in range(i + 1, len(similarity_matrix)):
        if similarity_matrix[i, j] > threshold:
            graph.add_edge(i, j, weight=similarity_matrix[i, j])

````

````python
from collections import Counter

# Extract product names and communities from the graph
node_data = [(node, graph.nodes[node]['community'], etsy_data.loc[node, 'Name']) for node in graph.nodes]

# Group product names by community
community_keywords = {}
for _, community, name in node_data:
    if community not in community_keywords:
        community_keywords[community] = []
    community_keywords[community].extend(name.split())

# Count keywords in each community
community_word_counts = {community: Counter(keywords) for community, keywords in community_keywords.items()}

# Identify low-popularity communities (small size, low sales/popularity)
community_stats = []
for community, word_count in community_word_counts.items():
    community_size = sum(word_count.values())
    popularity = sum(
        etsy_data.loc[node, 'Popularity_Score']
        for node, comm, _ in node_data if comm == community
    )
    community_stats.append((community, community_size, popularity))

# Sort communities by low popularity
low_popularity_communities = sorted(community_stats, key=lambda x: (x[1], x[2]))[:10]

# Extract keywords from low-popularity communities
low_popularity_keywords = {
    community: community_word_counts[community].most_common(10)
    for community, _, _ in low_popularity_communities
}

# Display results in a readable format
import pandas as pd

result_df = pd.DataFrame({
    "Community": [x[0] for x in low_popularity_communities],
    "Size": [x[1] for x in low_popularity_communities],
    "Popularity": [x[2] for x in low_popularity_communities],
    "Top Keywords": [", ".join([kw[0] for kw in low_popularity_keywords[x[0]]]) for x in low_popularity_communities]
})

display(result_df)
````
